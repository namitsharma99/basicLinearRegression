```python
import numpy as np
import pandas as pd
import csv
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (40.0, 20.0)

data = pd.read_csv("dummy_data.csv")
print(data.shape)
print(data)
```

```python
x = data['X-axis'].values
y = data['Y-axis'].values

x_bar = np.mean(x)
y_bar = np.mean(y)

print(x_bar)
print(y_bar)
```

```python
n = len(x)
print(n)
```

```python
# finding equation y = mx + c

num = 0
den = 0

for i in range(n) :
    num += (x[i] - x_bar) * (y[i] - y_bar)
    den += (x[i] - x_bar) ** 2
    
m = num / den
c = y_bar - m * x_bar

print(m)
print(c)
```

```python
# trying linear regression model

plt.rcParams.update({'font.size': 50})

#actual points
plt.scatter(x, y, color='red', label='actual scatter plot', s=150)

# for line y = mx + c
# translating variables x and y as u and v, as v = mu + c

u = x
v = m * u + c
plt.plot(u, v, color='green', label='predicted line')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()

```

```python
# checking the accuracy of this model using R^2 technique
# using line v = m * u + c

num = 0
den = 0

for i in range(n):
    # y axis predicted :
    v_p = m * u[i] + c
    num += (v_p - y_bar) ** 2
    den += (y[i] - y_bar) ** 2
    
print(num/ den)
```
